```python
import pydicom
import cv2 as cv
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
path = "C:/Users/Basti/Datasets/cocacoronarycalciumandchestcts-2/deidentified_nongated/"
# Charger le fichier DICOM
img_path = path +"1/1"+ "/IM-0001-0020.dcm"
print(img_path)
ds = pydicom.dcmread(img_path)

# Convertir en tableau numpy
img = ds.pixel_array.astype(np.float32)

# Normalisation (optionnel mais recommand√© pour l'affichage)
img = cv.normalize(img, None, 0, 255, cv.NORM_MINMAX)
img = img.astype(np.uint8)
print(img.shape)
new_img = cv.imwrite("img_test.png",img)
# Affichage avec OpenCV
cv.imshow("DICOM", img)
cv.waitKey(0)
cv.destroyAllWindows()
```
